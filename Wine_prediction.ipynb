# <u><center>Part 2: Creating and training a Linear Regression Model

## <i>Creating Lists for training and testing:

```python
x=iris.iloc[:,0:4]
y=iris.iloc[:,4]
x.shape,y.shape
```

```python
flower_mapping = {'setosa': 0,'versicolor': 1,'virginica':2}
iris["Species"] = iris["species"].map(flower_mapping)
```

## <i>Splitting data for training and testing: 

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

## <i>Creating the linear regression model:

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

## <i>Predictions from the model:

```python
prediction = model.predict(x_test)
predictions = model.predict(x_test)
plt.scatter(y_test,predictions, color="purple")
plt.show()
```

## <i>Checking Accuracy of the model:

```python
print(accuracy_score(y_test, prediction)*100)
```
